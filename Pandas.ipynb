### Adding a new column

```python
# Creating a new column
# Method 1
data_emp['total_amount'] = data_emp['income'] + data_emp['bonus']
```

```python
data_emp.head()
```

```python
# Method 2
# Points to note
# : means you are requesting total_amount for every row present in data
# 8:10 is the index number of columns from 8 to 10(excluding 10) so 8 = income, 9 = bonus. 
# axis = 1 means you are requesting sum horizontally.
# axis = 0 means you are requesting sum vertically.

data_emp['total_amount'] = data_emp.iloc[:,8:10].sum(axis=1) 
```

```python
data_emp.head()
```

```python
data_emp['total_amount'] = data_emp.iloc[:,8:10].sum(axis=0) 
```

```python
data_emp.head()
```

```python
# Droping a column
# Method 1
data_emp = data_emp.drop(columns=['total_amount'])
```

## Advance Filtering of Data

```python
# Multiple condition check in .loc() function
```

```python
# AND(&) operator inside the .loc() function
data_emp.loc[(data_emp['country'] == "Greater London") & (data_emp['city'] == "Millwall Ward")]
```

```python
# OR(|) operator inside the .loc() function
data_emp.loc[(data_emp['country'] == "Greater London") | (data_emp['city'] == "Millwall Ward")]
```

```python
# Mix condition check
data_emp.loc[(((data_emp['country'] == "Greater London") | (data_emp['city'] == "Millwall Ward")) & (data_emp['bonus'] > 100))]
```

```python
# Storing the result
filter_2 = data_emp.loc[(((data_emp['country'] == "Greater London") | (data_emp['city'] == "Millwall Ward")) & (data_emp['bonus'] > 100))]
filter_2
```

```python
# Also, we can store the data in file, but when you see your file the index have different values as per the result. 
filter_2.to_csv(r"C:\Users\risha\OneDrive\Desktop\filter_2.csv", index=True)
```

### Index resetting

```python
# Now if you want the index to be reset from 0 to n, you need to use .reset_index() function.
filter_2 = filter_2.reset_index()
filter_2
```

```python
filter_2.to_csv(r"C:\Users\risha\OneDrive\Desktop\filter_2.csv", index=True)
```

```python
# Use drop=True to drop the old index.
filter_2 = filter_2.reset_index(drop=True)
filter_2
```
